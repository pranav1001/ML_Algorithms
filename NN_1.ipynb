```
germany_maxtemp_by_state = germany.groupby('State')['Maximum Temperature'].max()
germany_mintemp_by_state = germany.groupby('State')['Minimum Temperature'].min()
germany_avgtemp_by_state = germany.groupby('State')['Temperature'].mean()
germany_maxhumidity_by_state = germany.groupby('State')['Relative Humidity'].max()
germany_minhumidity_by_state = germany.groupby('State')['Relative Humidity'].min()
germany_humidity_by_state = germany.groupby('State')['Relative Humidity'].mean()
germany_maxwindspeed_by_state = germany.groupby('State')['Wind Speed'].max()
germany_minwindspeed_by_state = germany.groupby('State')['Wind Speed'].min()
germany_windspeed_by_state = germany.groupby('State')['Wind Speed'].mean()
state=germany['State'].unique()

plt.rc('lines', linewidth = 2)
plt.figure(figsize=(15,8),dpi=100)
plt.plot(state,germany_maxtemp_by_state,'r',label="Statewise Max temp")
plt.plot(state,germany_mintemp_by_state,'b',label="Statewise Min temp")
plt.plot(state,germany_avgtemp_by_state,'purple',label="Statewise Min temp")
plt.plot(state,germany_humidity_by_state, color='green', linestyle='dashed', label="Statewise Average Humidity")
plt.plot(state,germany_maxhumidity_by_state, color='lawngreen', linestyle='dashed', label="Statewise Maximum Humidity")
plt.plot(state,germany_minhumidity_by_state, color='springgreen', linestyle='dashed', label="Statewise Minimum Humidity")
plt.plot(state,germany_windspeed_by_state, color='darkorange', linestyle='-.', label="Statewise Average Windspeed")
plt.plot(state,germany_maxwindspeed_by_state, color='maroon', linestyle='-.', label="Statewise Maximum Windspeed")
plt.plot(state,germany_minwindspeed_by_state, color='yellow', linestyle='-.', label="Statewise Minimum Windspeed")
plt.title("Statewise Weather Data of Germany")
plt.xlabel('STATE NAMES')
plt.ylabel('Values')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
